```python
# Function to parse text by speakers
def parse_text(text):
    # Regex to match speaker patterns more accurately
    pattern = r"\b((Mr|Ms|Senator)\s+[A-Z][A-Z]+)|(CHAIRMAN|CHAIRWOMAN)\b"
    
    # Find all occurrences of the pattern
    matches = list(re.finditer(pattern, text))
    
    # Extract paragraphs and corresponding speakers
    paragraphs = []
    speakers = []
    start = 0
    speaker = "Unknown"  # Initialize to handle text before first speaker if any
    
    for match in matches:
        end = match.start()
        if end != start:
            # Append text preceding current speaker
            paragraphs.append(text[start:end].strip())
            speakers.append(speaker)
        
        # Update speaker to include the title (if present) and name, and start position for next text
        speaker = match.group().strip()
        start = match.end()
    
    # Capture any remaining text after the last speaker
    if start < len(text):
        paragraphs.append(text[start:].strip())
        speakers.append(speaker)
    
    # Create a DataFrame
    df = pd.DataFrame({
        "Para_Num": range(1, len(paragraphs) + 1),
        "Speaker": speakers,
        "Text": paragraphs
    })
    
    return df
```
