# Covariate Analysis

This notebook verifies the rank-standardization results (`seniority_rs`, `abs_dwnom1_rs`) and plots the distribution of our key covariates.

```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")

# Load enriched sentences
df = pd.read_csv("../data/sentences_enriched.csv")
print(f"Loaded {len(df):,} sentences.")
df.head()
```

```
matched_df = df.dropna(subset=["bioguide_id"])
print(f"{len(matched_df):,} sentences matched to a legislator.")
```

## Rank Standardization Checks

The covariates `seniority_rs` and `abs_dwnom1_rs` should be distributed roughly uniformly between 0 and 1, though clustering can occur at certain seniority levels.

```
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(matched_df["seniority_rs"], bins=20, ax=axes[0])
axes[0].set_title("Distribution of seniority_rs")

sns.histplot(matched_df["abs_dwnom1_rs"], bins=20, ax=axes[1])
axes[1].set_title("Distribution of abs_dwnom1_rs")

plt.tight_layout()
plt.show()
```

## Vote Share Distribution

```
plt.figure(figsize=(8, 5))
sns.histplot(matched_df["vote_pct"].dropna(), bins=30)
plt.title("Distribution of Vote Share (vote_pct)")
plt.xlabel("Vote Share (%)")
plt.ylabel("Count (Sentences)")
plt.show()
```
